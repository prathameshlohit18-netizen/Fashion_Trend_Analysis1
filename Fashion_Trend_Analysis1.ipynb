# Brand and Category Analysis

# Top Brands

## Which brands have the most products

```python
brand_counts = df.Brand.value_counts().reset_index().head(10)
brand_counts.columns = ['Brand','Product_Count']

plt.figure(figsize=(15,8))
plt.bar(brand_counts["Brand"], brand_counts["Product_Count"], color=sb.color_palette("ch:start=.9,rot=-.7", len(brand_counts)),edgecolor='Black')
plt.title("Top 10 Brands by Number of Products")
plt.xlabel("Brand")
plt.ylabel("Number of Products")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Which categories dominate overall

```python
category_counts = df["Category"].value_counts().reset_index()
category_counts.columns = ["Category", "Product_Count"]

plt.figure(figsize=(14,7))
plt.pie(category_counts["Product_Count"], labels=category_counts["Category"], autopct='%1.1f%%', startangle=180,
           pctdistance=0.85,labeldistance=1.1,wedgeprops={'linewidth':1, 'edgecolor':'Black'},
           colors=sb.color_palette("vlag", len(category_counts)))
plt.title("Product Distribution across Category")
plt.show()
```

## which categories dominate in budget,mid,premium,luxury

```python
luxury=df.groupby(['Price_Segment','Category']).size().reset_index(name='Count').sort_values(['Price_Segment','Category'])
sb.barplot(data=luxury,x='Price_Segment',y='Count',hue='Category')
plt.xticks(rotation=90)
plt.show()
```

# Product Type Analysis

## Which product types are most common

```python
product_counts=df.Product_Name.value_counts().head(10)

plt.figure(figsize=(14,7))
product_counts.sort_values().plot(kind="pie", color=sb.color_palette("Spectral", len(product_counts)),
                                  )
plt.title("Top 15 Most Common Product Types")
plt.xlabel("Product Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

## compare product type distribution across genders

```python
productdistribution=df.groupby(['Gender','Product_Name']).size().reset_index(name='Count').sort_values(['Gender','Count'],ascending=[True,False])
productdistribution=productdistribution.groupby('Gender').head(5)
plt.figure(figsize=(15,8))
sb.barplot(y=productdistribution.Product_Name, x=productdistribution.Count,
             hue=productdistribution.Gender, dodge=True, palette="Blues",edgecolor='black')
plt.title("productdistribution.")
plt.xlabel("Count")
plt.ylabel("Product Type")
plt.legend(title="Gender",bbox_to_anchor=(1.0,0.9))
plt.tight_layout()
plt.show()


```

# Discount Analysis

## How many products fall into each discount tier

```python
discount_tier=df.groupby(['Discount_Tier']).size()
discount_tier
plt.figure(figsize=(8,7))
discount_tier.plot(
    kind='pie',
    autopct='%1.1f%%',
    wedgeprops={'edgecolor':'black'}
)
plt.legend(bbox_to_anchor=(1.35,1.0))
plt.ylabel("")
plt.show()
```

## which product types get the highest average discount

```python
top5_discounts = df.groupby('Product_Name')['Discount_Percentage'].mean().nlargest(5).reset_index(name='Discount')

plt.figure(figsize=(10,6))
sb.barplot(x='Discount', y='Product_Name', data=top5_discounts, palette='YlOrBr', edgecolor="black")

plt.title("Top 5 Products with Highest Average Discount", fontsize=14)
plt.xlabel("Average Discount Percentage")
plt.ylabel("Product Name")
plt.show()
```

## which categories rely most on heavy discounts

```python
avg_discount_category = df.groupby("Category")["Discount_Percentage"].mean().reset_index().sort_values("Discount_Percentage", ascending=False)

plt.figure(figsize=(12,6))
sb.violinplot(x=df.Category, y=df.Discount_Percentage, palette="coolwarm",edgecolor='black')

plt.title("Discount Reliance Across Categories")
plt.xlabel("Category")
plt.ylabel("Discount (%)")
plt.show()
```

## are higher-priced products more likely to have higher discounts

```python
df.groupby(['Category','Discount_Tier']).size().nlargest(5).reset_index(name='count').sort_values(['Category','Discount_Tier'])
```

```python
plt.scatter(x=df.Original_Price,y=df.Discount_Percentage,color="green",edgecolor="black")
plt.title("higher price product get higher discount")
plt.xlabel("Original_Price")
plt.ylabel("Discount_Percentage")
plt.tight_layout()
plt.show()
```

# Price Segment Analysis

## distribution of products across budget,mid,premium,luxury

```python
df['Price_Segment'].value_counts().reset_index(name='count').sort_values(['Price_Segment'])
```

```python
d = df.Price_Segment.value_counts()

colors = sb.color_palette("Set2") 

d.plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=colors,
    wedgeprops={'edgecolor':'black'}
)

plt.title("Price Segment Distribution")
plt.ylabel("") 
plt.show()
```

## which categories dominate in each segment

```python
k=df.groupby(['Price_Segment','Category']).size().reset_index(name='counts').sort_values(['Price_Segment','counts'],ascending=[True,False])
l=k.groupby('Price_Segment').head(5)
l
```

```python
plt.figure(figsize=[8,5])
sb.barplot(data=l, x='Price_Segment', y='counts', hue='Category')
plt.title("Price distribution according to categories")
plt.legend(bbox_to_anchor=(1.26, 1))
plt.tight_layout()
```

## compare brand positioning (budget vs premium  brands)

```python
brand_positioning = df.groupby(["Brand", "Price_Segment"]).size().reset_index(name="Count").sort_values(["Brand", "Count"], ascending=[True, False])
top_brands =brand_positioning.groupby("Brand")["Count"].sum().sort_values(ascending=False).head(10)   

plt.figure(figsize=(12,7))
sb.barplot(x=top_brands.index,y=top_brands.values,palette="flare",edgecolor='black')
plt.title("Top 10 Brands by Market Presence (Budget vs Premium brands)", fontsize=14)
plt.xlabel("Brand")
plt.ylabel("Number of Products")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
